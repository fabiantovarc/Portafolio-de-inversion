```python
# Clculo en Python del punto 2 (b): varianza y volatilidad mensual del portafolio
# Datos del enunciado
import numpy as np

mu = np.array([0.010, 0.006, 0.008])  # no se usa en este punto, pero lo dejo para referencia
Sigma = np.array([
    [0.0225, 0.0060, 0.0030],
    [0.0060, 0.0121, 0.0022],
    [0.0030, 0.0022, 0.0169]
])
w = np.array([0.50, 0.30, 0.20])

# Varianza del portafolio: sigma_p^2 = w' * Sigma * w
var_p = float(w.T @ Sigma @ w)

# Volatilidad mensual (desviacin estndar)
vol_p = np.sqrt(var_p)

var_p, vol_p
```
